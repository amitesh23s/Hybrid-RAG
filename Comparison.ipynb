```python
import matplotlib.pyplot as plt

categories = ['Comprehensiveness', 'Diversity', 'Empowerment', 'Directness']
hybrid_rag = [6, 3, 3, 5]
naive_rag = [1, 3, 1, 0]
ties = [0, 1, 3, 2]

x = range(len(categories))

plt.figure(figsize=(10, 6))
plt.bar(x, hybrid_rag, width=0.25, label='Hybrid RAG', align='center')
plt.bar([i + 0.25 for i in x], naive_rag, width=0.25, label='Naive RAG', align='center')
plt.bar([i + 0.5 for i in x], ties, width=0.25, label='Ties', align='center')

plt.xticks([i + 0.25 for i in x], categories)
plt.ylabel('Number of Wins')
plt.title('RAG Performance Comparison Across Categories')
plt.legend()
plt.tight_layout()
plt.show()
```
